```python
%matplotlib inline

import networkx as nx
```

A graph $G$ is an ordered pair $G = (V, E)$ comprising a set $V$ of vertices or nodes or points together with a set $E$ of edges or arcs or lines, which are 2-element subsets of $V$.

## Adding individual nodes

```python
G = nx.Graph()

G.add_node("a")
G.add_node("b")
G.add_node("c")

G.nodes()
```

## Adding nodes from a list

```python
G = nx.Graph()

G.add_nodes_from(["a", "b", "c"])

G.nodes()
```

## Basic graph visualization

```python
nx.draw(G,pos=nx.spring_layout(G))
```

## Adding nodes and edges

Note that networkx will create correspding nodes to edges of links  if they do not exist.

```python
G = nx.Graph()

G.add_edges_from([ ("a", "b"), ("b", "c") ])

nx.draw(G,pos=nx.spring_layout(G))
```

More information about graph creation can be found in the [networkx tutorial](https://networkx.readthedocs.io/en/stable/tutorial/tutorial.html). There is particulary a whole set of shortcut to access nodes and edges through loops.

```python
G = nx.fast_gnp_random_graph(20, 0.2)

nx.draw(G,pos=nx.spring_layout(G))
```
